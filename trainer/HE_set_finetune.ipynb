```python
import json
import os

# Define the 11 classes from 'raw_classification' in the HE dataset
classes = [
    'apoptotic_body',
    'ductal_epithelium',
    'fibroblast',
    'lymphocyte',
    'macrophage',
    'mitotic_figure',
    'myoepithelium',
    'neutrophil',
    'plasma_cell',
    'tumor',
    'vascular_endothelium'
]

# Sort the classes alphabetically to assign consistent IDs
classes.sort()

# Create the mapping dictionary: {ClassName: ID}
class_map = {label: idx for idx, label in enumerate(classes)}

# Display the map
print("HE Dataset Class Map:")
print(json.dumps(class_map, indent=4))

# Save the map to a JSON file in the same directory as this notebook
output_filename = 'he_class_map.json'
with open(output_filename, 'w') as f:
    json.dump(class_map, f, indent=4)

print(f"\nClass map successfully saved to: {os.path.abspath(output_filename)}")
```
